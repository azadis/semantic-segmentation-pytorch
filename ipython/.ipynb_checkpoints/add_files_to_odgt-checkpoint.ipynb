```python
import numpy as np
import glob
import os
from PIL import Image
```

```python
images = glob.glob('../data/ADE_bedroom/images/training/*.jpg')
train_txt = ''
for im in images:
#     if 'ADE_train_00003963' in im or 'ADE_train_00003677' in im or 'ADE_train_00004169' in im\
#         or 'ADE_train_00003598' in im or 'ADE_train_00003881' in im or 'ADE_train_00000371' in im\
#         or 'ADE_train_00003744' in im or 'ADE_train_00003216' in im or 'ADE_train_00000501' in im\
#         or 'ADE_train_00000296' in im or 'ADE_train_00003401' in im or 'ADE_train_00003494' in im\
#         or 'ADE_train_00003636' in im or 'ADE_train_00004223' in im:
#         continue
    annot = im.replace('images', 'annotations')
    annot = annot.replace('.jpg', '.png')
    img = Image.open(im)
    im = im.replace('../data/', '')
    annot = annot.replace('../data/', '')
    h,w = img.size
    train_txt += '{"fpath_img": "%s", "fpath_segm": "%s", "width": %s , "height": %s}\n'%(im, annot, w, h)
with open('../data/training_bedroom.odgt', 'w') as file:
    file.write(train_txt)
    
    
    
    
```

```python
images = glob.glob('../data/ADE_bedroom/images/validation/*.jpg')
val_txt = ''
for im in images:
    if 'ADE_train_00003963' in im or 'ADE_val_00001136' in im or 'ADE_val_00000120' in im:
        continue
    annot = im.replace('images', 'annotations')
    annot = annot.replace('.jpg', '.png')
    img = Image.open(im)
    im = im.replace('../data/', '')
    annot = annot.replace('../data/', '')
    h,w = img.size
    val_txt += '{"fpath_img": "%s", "fpath_segm": "%s", "width": %s , "height": %s}\n'%(im, annot, w, h)
with open('../data/validation_bedroom.odgt', 'w') as file:
    file.write(val_txt)
    
    
    
    
```
